# Python Set

## Definition
A **set** is an unordered collection of **unique elements** in Python.  
Sets are **mutable**, but they do **not allow duplicate values**.

---

## Syntax
```python
my_set = {1, 2, 3}

````python
fruite = {"apple", "bananna", "jackfruit", "pynaple", "orange"}

# add()
fruite.add("grapes")  # Add one element
print(fruite)
````

````python
# remove()
fruite.remove("apple")   # Remove element (error if missing)
print(fruite)
````

````python
x = {"apple", "bananna", "jackfruit", "pynaple", "orange"}
y = {"chilli", "onion", "coccumber", "apple"}

# union()

x.union(y) # Combine elements
````
