# Python Tuple

## Definition
A **tuple** is an ordered collection of elements in Python that is **immutable**, meaning once a tuple is created, its elements **cannot be changed, added, or removed**.

---

## Syntax
```python
my_tuple = (1, 2, 3)

````python
animale = ("lion", "tiger", "elephant", "deer", "cheetha", "gorilla")
print(animale)
````

````python
# count() method
numb = (1,2,3,4,5,5,6,7,5,8,9,4,3,7,9)  # hear "numb" is the variable it stores the tuple

numb.count(3)  # variable is connected to the function "count" as it counts number of value repeted in tuple
````

````python
# index() method
numb = (1,2,3,4,5,5,6,7,5,8,9,4,3,7,9)

numb.index(6) # as index count the position of the number,(the count starts from "0" )
````

````python
# min & max method
numb = (1,2,3,4,5,5,6,7,5,8,9,4,3,7,9)

# min(numb) # minimum
max(numb)   # maximum
````

# Python Dictionary

## Definition
A **dictionary** is a built-in data type in Python used to store data in **key–value pairs**.  
It is **unordered, mutable**, and keys must be **unique and immutable**.

---

## Syntax
```python
my_dict = {
    "name": "Nithish",
    "age": 22,
    "course": "Python"
}

````python
my_dict = {
    "name": "Nithish",       # in this ("name"= key ) & ("Nithish"= value)
    "age": 22,
    "course": "Python"
}

# get mentod
my_dict.get("name")   # Access value safely

# keys method
my_dict.keys()          # Get all keys

# values method
my_dict.values()          # Get all values

# items method
my_dict.items()           # Get key–value pairs
````

````python
# pop() method

my_dict.pop("name") # Remove specific key
print(my_dict)

# popitem() method

my_dict.popitem()  #Remove last item
````

````python
# update() method

my_dict.update({"gender" : "male"})   # Add/update items

print(my_dict)
````

````python
# clear() method

my_dict.clear()   # Remove all items

print(my_dict)
````

````python
# fromkeys()

keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)   # Creates a new dictionary from given keys with a common value
print(new_dict)
````

````python
new_dict.setdefault("d", 4)   # Returns the value of a key. If the key does not exist, inserts it with a default value.
````
