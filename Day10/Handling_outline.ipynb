````python
import numpy as np

data = [1, 2, 3, 4, 5, 6, 9, 10, 12, 55, 60]
data
````

````python
Q1 = np.percentile(data,25)
````

````python
Q1
````

````python
Q3 = np.percentile(data,75)
````

````python
Q3
````

````python
Q1 + Q3
````

````python
IQR = Q3 - Q1
````

# Formula:
 ### Lower limit  = Q1 âˆ’ 1.5 Ã— IQR
 ### Upper limit  = Q3 + 1.5 Ã— IQR

````python
low = Q1 - 1.5 * IQR
upp = Q3 + 1.5 * IQR
````

````python
low,upp
````

````python
Outliers = []
````

````python
data
````

````python
#  Detect Outliers Using Loop

for i in data:
    if ((i > upp) or (i < low)):
        Outliers.append(i)
````

````python
low  = 7.75
upp  = 22.25
````

````python
# Remove Outliers (List Comprehension)

list_ = [i for i in data if i not in Outliers]
list_
````
