## Python Loops

A **loop** in Python is used to **repeat a block of code multiple times** until a condition is met or until all items in a sequence are processed.

Python mainly uses two types of loops:
- `for` loop
- `while` loop

---

## 1. `for` Loop

A `for` loop is used to **iterate over a sequence** such as a list, tuple, string, or range.

### Syntax
```python
for variable in sequence:
    statement(s)

### Working of `for` Loop in Python (Flowchart Explanation)

A `for` loop in Python is used to **iterate(repitation) through items in an iterable** such as a list, tuple, string, or range.

The loop automatically moves from the **first item to the last item**, executing the given statement for each item.

---

### Step-by-step Working (as shown in the diagram)

1. **First item in iterable**
   - The loop starts by picking the **first element** from the iterable.

2. **Execute expression**
   - The code inside the `for` loop is executed for the current item.

3. **Get next item**
   - After execution, Python automatically moves to the **next item** in the iterable.

4. **Check: Is this the last item?**
   - If **NO** → repeat the loop  
   - If **YES** → exit the loop

5. **End for-loop**
   - The loop stops after the **last item is processed**

---

### Example Code
```python
subjects = ["SUB001", "SUB002", "SUB003"]

for sub in subjects:
    print(sub)

## 2.`while` Loop
A while loop executes as long as a condition is True.

### Syntax
```python
while condition:
    statment(s)

### Working of `while` Loop in Python (Flowchart Explanation)

A `while` loop in Python is used to **repeat a block of code as long as a given condition is True**.  
Unlike a `for` loop, the number of iterations is **not fixed in advance**.

---

### Step-by-step Working (as shown in the diagram)

1. **Enter while loop**
   - Python enters the `while` loop and reaches the condition check.

2. **Test Expression**
   - The condition is evaluated.
   - Example: `count <= 5`

3. **If the condition is True**
   - The statements inside the loop are executed.
   - After execution, control goes back to the **test expression** again.

4. **If the condition is False**
   - The loop stops.
   - Python exits the while loop.

5. **Exit while loop**
   - Execution continues with the next statement after the loop.

---

### Example Code
```python
count = 1

while count <= 3:
    print(count)
    count += 1

````python
num = 1

while num < 20:
    print(num)
    num += 3
````
