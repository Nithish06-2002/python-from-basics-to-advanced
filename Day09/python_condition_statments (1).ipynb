# ðŸ Python Control Statements

Python **control statements** are used to **control the flow of execution** of a program based on conditions.

## What are Control Statements?
Control statements decide:
- **Which block of code should run**
- **When it should run**
- **Whether it should run or not**

They are mainly used for **decision making** and **flow control**.

## 1. `if` Statement

The `if` statement runs a block of code **only if the condition is True**.

```python
x = 10

if x > 5:
    print("x is greater than 5")

# ðŸ Python `if`, `elif`, and `else` Statements

In Python, `if`, `elif`, and `else` are **decision-making (control) statements**.  
They are used to execute code **based on conditions**.

## 1. `if` Statement

The `if` statement runs a block of code **only when the condition is True**.

```python

## 2ï¸âƒ£ if â€“ elif statement (multiple conditions, no else)


````python
anusha, arvind = 80, 62

if anusha > arvind:
    print("arvind scores greater than anusha")
elif anusha == arvind:
    print("Both scores equally")
````

### What kind of logic this is:

- **Multiple condition checking**
- Conditions are checked top to bottom
- Only the first `True` condition runs
- No default (`else`) case

## 3ï¸âƒ£ if â€“ elif â€“ else statement 


````python
if arvind > anusha:
    print("arvind scores greater than anusha")
elif anusha == arvind:
    print("Both scores equally")
else:
    print("anusha scores greater than arvind")
````

### What kind of logic this is:

- Complete **decision-making** structure
- Covers all possible cases
- Exactly one block executes

## 4ï¸âƒ£ NESTED ifâ€“else

````python
marks = 28                     # assigning the value 28 to the variable marks

if marks > 35:                 # checks if marks is greater than 35
    print("He is Pass")        # runs only if the above condition is True

    if marks > 60:             # nested if: checks if marks is greater than 60
        print("He is in Distinction")  # runs if marks > 60 is True
    else:                      # runs if marks > 60 is False
        print("He is in Mid Range")    # student is neither distinction nor fail

else:                          # runs when marks > 35 condition is False
    print("Sadly, He is Fail") # student has failed
````

### Key idea (inside the code logic)
-Outer `ifâ€“else` â†’ Pass or Fail
-Inner `ifâ€“else` â†’ Distinction or Mid Range
-Inner block runs only `if` the outer condition is True

````python
### example

cp_score = 80                    # assigning company performance score

if cp_score > 30:                 # checks if score is above minimum
    if cp_score > 50:             # checks for high performance
        print("qualified")        # high score
    else:
        print("try again")        # mid-range score
else:
    print("disqualified")         # very low score
````
