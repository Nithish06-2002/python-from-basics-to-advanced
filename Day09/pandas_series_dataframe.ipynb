# Pandas Series

````python
import pandas as pd
````

````python
# series method
series = pd.Series([1,2,3,4,5,6,7])
````

````python
print(series)
````

````python
# series method with index
series1 = pd.Series([1,2,3,4,5], index = ['a','s','e','f','c'])
print(series1)
````

````python
# Accesing index

print(series1['a'])
print(series1['f'])
````

import pandas as pd

````python
# DataFrame method using list method

data = pd.DataFrame([1,2,3,4,5,6])
print(data)
````

````python
# DataFrame method using dictionary method

data = {
    'Name': ['Asha', 'Ravi', 'John'],
    'Age': [23, 25, 30],
    'Score': [85, 90, 88]
}

df = pd.DataFrame(data)
df
````

````python
data = pd.DataFrame(
    {'Name': ['Tom', 'Jack', 'Steve', 'Ricky', 'Sam', 'Will', 'Root'],
        'Age': [28, 34, 29, 42, 17, 17, 45]},
    index=['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7'])

data.head()
````

````python
# THIS STEP IS USED TO SELECT THE COLUME
data['Name']
data['Age']
````

````python
data.head(10)
````

````python
# row selection

data.loc['C1','Name']    # it is used to just to retrive the data like by menctioning the lables
data.iloc[2,1]             # in this it works like by menctioning the position in the table
````
