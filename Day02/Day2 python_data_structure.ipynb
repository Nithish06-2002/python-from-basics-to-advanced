# üêç Python Strings & String Methods 

## What is a String in Python?
- A string is a **sequence of characters**
- Strings are written inside **single (`'`) or double (`"`) quotes**
- Strings are **immutable** (cannot be changed in place)

```python
text = "hello"

````python
# capitalize() ‚Üí makes the first character uppercase
text = "hello world"
result = text.capitalize()
print(result)        # Output: Hello world
````

````python
# isalpha() ‚Üí checks if the string contains only letters (no spaces, no numbers)
text = "helloworld"
result = text.isalpha()
print(result)           # Output: True
````

````python
# isnumeric() ‚Üí checks if the string contains only numbers
text = "123456" # if it contain the 123456A the statment will be false because it consly recognises the number not alphabetes
result = text.isnumeric()
print(result)         # Output: True
````

````python
# isupper() ‚Üí checks whether all letters in the string are uppercase
text = "hello world"
result = text.isupper()
print(result)           # Output: False
````

````python
# split() ‚Üí splits the string into a list using space by default
text = "Hi Alex" 
result = text.split()
print(result)             # Output: ['Hi', 'Alex']
````

````python
# title() ‚Üí converts the first letter of each word to uppercase
text = "hello world"
print(text.title())             # Output: Hello World
````

````python
# strip() ‚Üí removes extra spaces from the beginning and end of the string
text = "   Hello   nithish           how are you  "
print(text.strip())             # Output: Hello
````

````python
# replace() ‚Üí replaces one character or word with another
text = "a b c"
result = text.replace('a', 'd')
print(result)   # Output: d b c
````
